```python
# Import necessary libraries
from IPython.display import display, clear_output
import ipywidgets as widgets

class StoryNode:
    def __init__(self, text, choices):
        self.text = text
        self.choices = choices

class Game:
    def __init__(self):
        self.story = self.create_story()
        self.current_node = self.story
        self.player_choices = []
        self.group_votes = []

    def create_story(self):
        # Define the end node
        end_node = StoryNode("Your adventure ends here. Thanks for playing!", [])

        # Create nodes for the story
        choice_6 = StoryNode("You discover an ancient temple. Do you enter?", [
            StoryNode("Enter the temple.", [end_node]),
            StoryNode("Leave it alone.", [end_node]),
            StoryNode("Search around.", [end_node])
        ])

        choice_5 = StoryNode("A wise man offers potions. Which do you choose?", [
            StoryNode("Strength potion.", [choice_6]),
            StoryNode("Wisdom potion.", [choice_6]),
            StoryNode("Invisibility potion.", [choice_6])
        ])

        choice_4 = StoryNode("A fork in the road. Go left or right?", [
            StoryNode("Left to the mountains.", [choice_5]),
            StoryNode("Right to the valley.", [choice_5]),
            StoryNode("Go back.", [end_node])
        ])

        choice_3 = StoryNode("A dragon appears! What do you do?", [
            StoryNode("Fight!", [choice_4]),
            StoryNode("Flee.", [choice_4]),
            StoryNode("Negotiate.", [choice_4])
        ])

        choice_2 = StoryNode("You find a map. What now?", [
            StoryNode("Follow it.", [choice_3]),
            StoryNode("Ignore it.", [choice_3]),
            StoryNode("Burn it.", [end_node])
        ])

        choice_1 = StoryNode("You awaken in a strange land. What next?", [
            StoryNode("Explore.", [choice_2]),
            StoryNode("Seek help.", [choice_2]),
            StoryNode("Camp.", [end_node])
        ])

        # Starting point
        start_node = StoryNode("You find yourself in a lush, unfamiliar land.", [
            choice_1,
        ])

        return start_node

    def display_choices(self):
        clear_output(wait=True)
        print(self.current_node.text)

        if self.current_node.choices:
            buttons = []
            for i, choice in enumerate(self.current_node.choices):
                button = widgets.Button(description=f"{i + 1}. {choice.text}")
                button.on_click(lambda _, index=i: self.make_choice(index))
                buttons.append(button)
                display(button)
        else:
            self.display_group_votes()

    def make_choice(self, index):
        self.player_choices.append(self.current_node.choices[index].text)
        self.current_node = self.current_node.choices[index]

        if not self.current_node.choices:  # If it's an end node, display the final outcome
            clear_output(wait=True)
            print(self.current_node.text)
            self.record_vote()
        else:
            self.display_choices()

    def record_vote(self):
        print("\nVoting for the next choice...")
        vote_buttons = []
        for i, choice in enumerate(self.player_choices):
            button = widgets.Button(description=f"Vote for: {choice}")
            button.on_click(lambda _, index=i: self.cast_vote(index))
            vote_buttons.append(button)
            display(button)

    def cast_vote(self, index):
        self.group_votes.append(self.player_choices[index])
        clear_output(wait=True)
        print(f"You voted for: {self.player_choices[index]}")
        self.display_group_votes()

    def display_group_votes(self):
        print("Group Votes:")
        for vote in set(self.group_votes):
            print(f"{vote}: {self.group_votes.count(vote)} votes")
        print("\nThank you for participating!")

    def play(self):
        self.display_choices()

# Start the game
game = Game()
game.play()
```
